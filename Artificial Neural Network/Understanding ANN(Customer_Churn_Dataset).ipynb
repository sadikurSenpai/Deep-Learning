```python
df = pd.read_csv('/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv')
df
```

```python
df.info()  # checking non-null
```

```python
df.duplicated().sum() # checking duplicate values
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'], inplace=True) # removing unnecessary columns
df
```

```python
df.value_counts(['Geography']) # checking the geography column's categorical values
```

```python
df.value_counts(['Gender']) # checking the gender column's categorical values
```

```python
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)
df
```

```python
df = df.astype(int)
```

```python
df
```

```python
X = df.drop(columns=['Exited'])
X
```

```python
y = df['Exited']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
X_test_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(11,activation='sigmoid',input_dim=11))
model.add(Dense(11,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = y_pred.argmax(axis=-1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)  
```
